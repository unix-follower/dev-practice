```python
from foodb_helper import read_food_data, prepare_dataset

original_food_df = read_food_data()
original_food_df
```

```python
food_df = original_food_df
food_df, x, y = prepare_dataset(food_df)
food_df
```

```python
display(food_df.dtypes)
display(food_df.describe())
food_df["food_type"].value_counts()
```

```python
import sklearn
from sklearn.preprocessing import OrdinalEncoder
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

print(sklearn.__version__)

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

pipeline = Pipeline([
    ("oe", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)),
    ("scaler", StandardScaler()),
    ("svc", SVC())
])
pipeline
```

```python
pipeline.fit(x_train, y_train).score(x_test, y_test)
```

```python
import pandas as pd

food_name = "Herbal tea"
# food_name = "Fish oil"
# food_name = "Taco shell"
# food_name = "Tostada shell"
input_data = pd.DataFrame({
    "name": food_name,
    "description": "data is not specified",
    "category": "unknown",
}, index=[0])
display(input_data.shape)
pipeline.predict(input_data)
```
