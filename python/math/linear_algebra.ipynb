#### Page 3. 1.3 Example.
$$
\begin{gather}
3x_1+2x_2=7 \\
-x_1+x_2=6
\end{gather}
$$

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import sympy as sym
from sympy.matrices import Matrix
import scipy
import tensorflow as tf

print(f"SymPy version: {sp.__version__}")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {matplotlib.__version__}")
print(f"Seaborn version: {sns.__version__}")
print(f"TensorFlow version: {tf.__version__}")
```

```python
coefficient_matrix = np.array([
    [3, 2],
    [-1, 1],
])
print(f"coefficient_matrix shape={coefficient_matrix.shape} ndim={coefficient_matrix.ndim}")

constant_vector = np.array([7, 6])
print(f"constant_vector shape={constant_vector.shape} ndim={constant_vector.ndim}")

result = np.linalg.solve(coefficient_matrix, constant_vector)
print(type(result))
print(f"result shape={result.shape} ndim={result.ndim}")
print(result)

x1, x2 = result
assert -1 == x1, f"The actual value: {x1}"
assert 5 == x2, f"The actual value: {x2}"

figure, axes = plt.subplots()
axes: plt.Axes
axes.plot(result, marker="o", markersize=5)
axes.set_xlim(0, 1.5)
axes.set_ylim(0, 5.5)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_xticks(np.arange(0, 3))
axes.set_yticks(np.arange(0, 6))
plt.show()
axes.clear()

sns.lineplot(result, marker="o", markersize=5)
plt.show()

sns.scatterplot(result, marker="x")
plt.show()
```

```python
coefficient_matrix = [
    [3, 2],
    [-1, 1],
]
constant_vector = [7, 6]

result = scipy.linalg.solve(coefficient_matrix, constant_vector)
print(type(result))
print(f"result shape={result.shape} ndim={result.ndim}")
print(result)
x1, x2 = result
assert -1 == x1
assert 5 == x2
```

```python
x1, x2 = sym.symbols("x1, x2")

equation1 = 3 * x1 + 2 * x2 - 7
equation2 = -x1 + x2 - 6

sym.solve([equation1, equation2], (x1, x2))
```

```python
coefficient_matrix = tf.constant([
    [3, 2],
    [-1, 1]
], dtype=tf.float32)
print(f"coefficient_matrix shape={coefficient_matrix.shape} ndim={coefficient_matrix.ndim}")

constant_vector = tf.constant([7, 6], dtype=tf.float32)
print(f"constant_vector shape={constant_vector.shape} ndim={constant_vector.ndim}")
constant_vector = tf.reshape(constant_vector, (-1, 1))  # Reshape to a column vector
print(f"constant_vector shape={constant_vector.shape} ndim={constant_vector.ndim}")

result = tf.linalg.solve(coefficient_matrix, constant_vector)
result = tf.squeeze(result)

print(type(result))
print(f"result shape={result.shape} ndim={result.ndim}")
print(result)

x1, x2 = result
assert tf.math.equal(x1, -1)
assert tf.math.equal(x2, 5)
```

#### Page 3. 1.4 Example.
$$
\begin{gather}
3x_3=9 \\
x_1+5x_2-2x_3=2 \\
\frac{1}{3}x_1+2x_2=3
\end{gather}
$$

```python
x1, x2, x3 = sym.symbols("x1, x2, x3")

equation1 = 3 * x3 - 9
equation2 = x1 + 5 * x2 - 2 * x3 - 2
equation3 = 1/3 * x1 + 2 * x2 - 3

result: dict = sym.solve([equation1, equation2, equation3], (x1, x2, x3))
print(type(result))
print(result)

assert 3 == result[x1], f"The actual value: {result[x1]}"
assert 1 == result[x2], f"The actual value: {result[x2]}"
assert 3 == result[x3], f"The actual value: {result[x3]}"
```

```python
coefficient_matrix = np.array([
    [3, 0, 0],
    [1, 5, -2],
    [1 / 3, 2, 0]
])
print(f"coefficient_matrix shape={coefficient_matrix.shape} ndim={coefficient_matrix.ndim}")

constant_vector = np.array([9, 2, 3])
print(f"constant_vector shape={constant_vector.shape} ndim={constant_vector.ndim}")

result = np.linalg.solve(coefficient_matrix, constant_vector)
print(type(result))
print(f"result shape={result.shape} ndim={result.ndim}")
print(result)

x1, x2, x3 = result
assert 3 == int(round(x1)), f"The actual value: {x1}"
assert 1 == int(round(x2)), f"The actual value: {x2}"
assert 3 == int(round(x3)), f"The actual value: {x3}"

figure, axes = plt.subplots()
axes: plt.Axes
axes.plot(result, marker="o", markersize=5)
axes.set(
    xlim=(0, 3.5),
    ylim=(0, 3.5),
    xticks=np.arange(0, 4),
    yticks=np.arange(0, 4),
    xlabel="x",
    ylabel="y"
)
plt.show()
axes.clear()

sns.lineplot(result, marker="o", markersize=5)
```

#### Page 5. 1.7 Example.
$$
\begin{gather}
x+y=0 \\
2x-y+3z=3 \\
x-2y-z=3
\end{gather}

```python
x, y, z = sym.symbols("x, y, z")

equation1 = x + y - 0
equation2 = 2 * x - y + 3 * z - 3
equation3 = x - 2 * y - z - 3

result: dict = sym.solve([equation1, equation2, equation3], (x, y, z))
print(type(result))
print(result)

assert 1 == result[x], f"The actual value: {result[x]}"
assert -1 == result[y], f"The actual value: {result[y]}"
assert 0 == result[z], f"The actual value: {result[z]}"
```

```python
x, y, z = sym.symbols("x, y, z")

equation1 = x + 2 * y - z - 1
equation2 = 2 * x + 4 * y - 2
equation3 = y - 3 * z

result: dict = sym.solve([equation1, equation2, equation3], (x, y, z))
print(type(result))
print(result)

assert 1 == result[x], f"The actual value: {result[x]}"
assert 0 == result[y], f"The actual value: {result[y]}"
assert 0 == result[z], f"The actual value: {result[z]}"
```
